```python
import matplotlib.pyplot as plt
import numpy as np

# ==========================================
# 0. 參數設定區 (Configuration) - 修改這裡即可
# ==========================================

# --- 字體大小 (Font Sizes) ---
FONT_MAIN_TITLE   = 40      # 整張圖的最上方大標題
FONT_SUB_TITLE    = 26      # 子圖標題 (如 Capital by Exchange)
FONT_PIE_PCT      = 18      # 圓餅圖裡面的百分比數字 (100%)
FONT_LEGEND_TITLE = 18      # 圖例的標題 (Asset Breakdown)
FONT_LEGEND_BODY  = 16      # 圖例的內容文字

# --- 佈局與距離 (Layout & Padding) ---
PAD_MAIN_TITLE    = 0.96    # 大標題的垂直位置 (越接近 1 越上面)
PAD_SUB_TITLE     = 25      # 子圖標題與圖表的距離
PCT_DISTANCE      = 0.82    # 圓餅內數字離圓心的距離 (0~1)

# ==========================================
# 1. 數據準備
# ==========================================
exchange_data = {
    'Bitget': 32748.57,
    'Binance': 26473.97,
    'Bybit': 29697.39,
    'Hyperliquid': 7998.65
}

symbol_data = {
    'Others (64 coins)': 39.1,
    'BEAT': 7.8,
    'PIPPIN': 7.7,
    'TRUTH': 5.0,
    'NIGHT': 4.9,
    'ICNT': 4.4,
    'ZEC': 4.3,
    'ORDI': 3.9,
    'LRC': 3.6,
    'TNSR': 3.2,
    'HYPE': 3.2,
    'SOMI': 3.0,
    'PIEVERSE': 2.7,
    'STABLE': 2.5,
    'RESOLV/USDT': 2.5,
    'RIVER/USDT': 2.3
}

# ==========================================
# 2. 繪圖初始化
# ==========================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(26, 13)) 
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS', 'SimHei', 'sans-serif']
plt.rcParams['axes.unicode_minus'] = False

# 設定全域大標題 (使用上面定義的變數)
fig.suptitle(
    '2025-12-16 資金分佈概覽 (Portfolio Distribution)', 
    fontsize=FONT_MAIN_TITLE, 
    weight='bold', 
    y=PAD_MAIN_TITLE
)

# ==========================================
# 3. 顏色策略
# ==========================================
# 左圖顏色
colors_exch = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'] 

# 右圖顏色 (Tab20 高對比)
labels_r = list(symbol_data.keys())
values_r = list(symbol_data.values())
cmap = plt.get_cmap("tab20")
raw_colors = cmap(np.linspace(0, 1, len(values_r)))

final_colors_r = []
for label, color in zip(labels_r, raw_colors):
    if "Others" in label:
        final_colors_r.append('#D3D3D3') # Others 固定灰色
    else:
        final_colors_r.append(color)

# ==========================================
# 4. 通用繪圖函數
# ==========================================
def draw_pro_pie(ax, values, labels, colors, title):
    # 1. 繪製圓餅
    wedges, texts, autotexts = ax.pie(
        values,
        labels=None, 
        autopct=lambda p: f'{p:.1f}%' if p > 2 else '', 
        startangle=90,
        colors=colors,
        pctdistance=PCT_DISTANCE, 
        wedgeprops=dict(width=0.4, edgecolor='white', linewidth=2)
    )
    
    # 2. 設定圓餅內數字字體
    for text, wedge in zip(autotexts, wedges):
        text.set_fontsize(FONT_PIE_PCT)  # <--- 使用變數
        text.set_weight('bold')
        
        # 智慧顏色調整
        facecolor = wedge.get_facecolor()
        if sum(facecolor[:3]) > 1.5: 
            text.set_color('#333333')
        else:
            text.set_color('white')

    # 3. 設定子標題
    ax.set_title(
        title, 
        fontsize=FONT_SUB_TITLE,      # <--- 使用變數
        pad=PAD_SUB_TITLE,            # <--- 使用變數
        weight='bold', 
        color='#333333'
    )

    # 4. 製作側邊圖例
    total = sum(values)
    legend_text = []
    for l, v in zip(labels, values):
        pct = (v / total) * 100
        legend_text.append(f'{l}  ({pct:.1f}%)')

    ax.legend(
        wedges, 
        legend_text,
        title="Asset Breakdown",
        title_fontsize=FONT_LEGEND_TITLE, # <--- 使用變數
        loc="center left",
        bbox_to_anchor=(1, 0, 0.5, 1), 
        fontsize=FONT_LEGEND_BODY,        # <--- 使用變數
        frameon=False,
        labelspacing=0.6
    )

# ==========================================
# 5. 執行繪圖
# ==========================================

# 繪製左圖
draw_pro_pie(
    ax1, 
    list(exchange_data.values()), 
    list(exchange_data.keys()), 
    colors_exch, 
    "Capital by Exchange"
)

# 繪製右圖
draw_pro_pie(
    ax2, 
    values_r, 
    labels_r, 
    final_colors_r, 
    "Top 15 Symbols Allocation"
)

# ==========================================
# 6. 輸出
# ==========================================
plt.tight_layout(rect=[0, 0, 0.95, 0.93]) 
plt.show()
```

```python
def coin_analysis(coin_porfolio_df, top_n=20):
    """Plot top N coins showing waterfall breakdown: base PnL, fees, and spread impact."""
    import matplotlib.pyplot as plt
    import numpy as np
    from matplotlib.patches import Patch
    
    # Calculate cumulative components
    cumulative_base = coin_porfolio_df[0].sum(axis=0)
    cumulative_no_spread = coin_porfolio_df[1].sum(axis=0)
    cumulative_total = coin_porfolio_df[2].sum(axis=0)
    
    fee_impact = cumulative_no_spread - cumulative_base
    spread_impact = cumulative_total - cumulative_no_spread
    
    # Select and sort top N
    top_symbols = cumulative_total.abs().nlargest(top_n).index
    sorted_idx = cumulative_total[top_symbols].sort_values(ascending=False).index
    
    print(f"\nComponent Breakdown for Top {top_n} Coins:")
    print(f"{'Symbol':<12} {'Base':>8} {'Fees':>8} {'Spread':>8} {'Total':>8}")
    print("="*50)
    for sym in sorted_idx:
        print(f"{sym:<12} {cumulative_base[sym]*100:>7.2f}% {fee_impact[sym]*100:>7.2f}% "
              f"{spread_impact[sym]*100:>7.2f}% {cumulative_total[sym]*100:>7.2f}%")
```
